# **1.Defanging an IP Address**

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".","[.]")

add = Solution()
print(add.defangIPaddr("1.1.1.1"))
print(add.defangIPaddr("255.100.50.0"))
```

# **2.Find Numbers with Even Numbers of Digits**

```python
from typing import List

class Solution:
    def findNumbers(self, nums: List[int]) -> List[int]:
        even_num = []
        for num in nums:
            if len(str(num)) % 2 == 0:
                even_num.append(num)
        return even_num
even = Solution()
print(even.findNumbers([12,345,2,6,7896]))
print(even.findNumbers([555,901,482,1771]))
```

# 3.Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        count=0
        freq = {}

        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1
        return count

result = Solution()
print(result.numIdenticalPairs([1,2,3,1,1,3]))
```

# 4.How Many Numbers Are Smaller Than the Current Number.

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        sorted_nums = sorted(nums)
        result = []

        for num in nums:
            result.append(sorted_nums.index(num))
        return result

result = Solution()
print(result.smallerNumbersThanCurrent([8,1,2,2,3]))
```

# 5.Subtract the Product and Sum of Digits of an Integer.

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        total = 0

        for d in str(n):
            product *= int(d)
            total += int(d)

        return product -total
result = Solution()
print(result.subtractProductAndSum(234))
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0

        for i in range(n):
            result ^= start + 2 * i

        return result
result = Solution()
print(result.xorOperation(5,0))
```
